### Task 1

```python
freq = 30
amp = 10
phase = 0.5
samples = 1000
duration = 0.1

freq1 = 20
amp1 = 10
phase1 = 0.5

freq2 = 30
amp2 = 20
phase2 = 0.5

freq3 = 30
amp3 = 10
phase3 = 1

t = np.linspace(0, duration, int(samples * duration), endpoint=False)

# A Sin(2*pi*f*t + phase)
y = amp * np.sin(2 * np.pi * freq * t + phase)
y1 = amp1 * np.sin(2 * np.pi * freq1 * t + phase1)
y2 = amp2 * np.sin(2 * np.pi * freq2 * t + phase2)
y3 = amp3 * np.sin(2 * np.pi * freq3 * t + phase3)


plt.figure(figsize=(8,4))
plt.plot(t, y, label=f"{freq, amp, phase}")
plt.plot(t, y1, label=f"{freq1, amp1, phase1}")

plt.title(f"Original vs Case 1")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### Task 3 (Sampling)

```python
freq = 10
duration = 1
t = np.linspace(0, duration, 1000)
y = np.cos(2 * np.pi * freq * t)

plt.figure(figsize=(8, 4))
plt.plot(t, y, label='Original Signal')
plt.grid()
plt.legend()
plt.show()
```

```python
fs = 5
ts = 1/fs

pulse_train = np.arange(0, duration, ts)
xt_sampled = np.cos(2 * np.pi * freq * pulse_train)
x_rs, t_rs = scipy.signal.resample(xt_sampled, 100, pulse_train)


# plt.stem(pulse_train, np.ones(len(pulse_train)))
```

```python
plt.figure(figsize=(14,10))

plt.subplot(4, 1, 1)
plt.stem(pulse_train, np.ones(len(pulse_train)))
plt.title('Sampling')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.tight_layout()

plt.subplot(4, 1, 2)
plt.stem(pulse_train, np.ones(len(pulse_train)))
plt.plot(t, y, 'r')
plt.title('Original Signal with Sampled points')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.tight_layout()


xt_sampled = np.cos(2 * np.pi * freq * pulse_train)
plt.subplot(4, 1, 3)
plt.stem(pulse_train, xt_sampled)
plt.title('Sampled Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.tight_layout()


plt.subplot(4, 1, 4)
plt.plot(t_rs, x_rs)
plt.plot(t, y)
plt.title('Reconstructed Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.tight_layout()


plt.show()
```

### Task 4 (Resampling)

#### Downsampling

```python
from scipy.signal import resample_poly

freq = 200
sampling_rate = 1000
t = np.linspace(0, 1, sampling_rate, endpoint=False)
x = np.sin(2 * np.pi * freq * t)

x_down = resample_poly(x, up=1, down=4)
t_down = np.linspace(0, 1, len(x_down), endpoint=False)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Original Signal (1000 Hz)')
plt.plot(t_down, x_down, 'o-', label='Downsampled (250 Hz)')
plt.title("Downsampling")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

#### Upsampling

```python
from scipy.signal import resample_poly

freq = 5
sampling_rate = 100
t = np.linspace(0, 1, int(sampling_rate * 0.1), endpoint=False)
x = np.sin(2 * np.pi * freq * t)

x_down = resample_poly(x, up=4, down=1)
t_down = np.linspace(0, 1, len(x_down), endpoint=False)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Original Signal (100 Hz)')
plt.plot(t_down, x_down, 'o-', label='Upsampled (400 Hz)')
plt.title("Upsampling")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### Task 5 (Filtering)

```python
sampling_rate_fs = 1000
duration = 0.1

freq1 = 2
freq2 = 15
freq3 = 30

t = np.linspace(0, 1, sampling_rate_fs, endpoint=False)
y1 = np.sin(2 * np.pi * freq1 * t)
y2 = np.sin(2 * np.pi * freq2 * t)
y3 = np.sin(2 * np.pi * freq3 * t)
```

```python
y = y1 + y2 + y3
```

#### Low pass filter

```python
from scipy.signal import butter, filtfilt

low_cutoff = 2

order = 4
nyq = 0.5 * sampling_rate_fs

normal_low_cutoff = low_cutoff / nyq

b, a = butter(4, normal_low_cutoff, btype='lowpass')

filtered_data1 = filtfilt(b, a, y)

plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Original Signal')
plt.plot(t, filtered_data1, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Low-Pass Filtering')
plt.legend()
plt.grid()
plt.show()
```

```python
print("Filtered signal max: ", np.max(filtered_data1))
print("Filtered signal min: ", np.min(filtered_data1))
print("Filtered signal mean: ", np.mean(filtered_data1))
```

```python
from scipy.fft import fft, fftfreq, rfft, rfftfreq

N = sampling_rate_fs

yf = fft(y)
xf = fftfreq(N, 1 / sampling_rate_fs)
filtered1 = fft(filtered_data1)

plt.figure(figsize=(10,6))
plt.plot(xf, np.abs(yf))
plt.plot(xf, np.abs(filtered1))
plt.xticks([-100, 0, 100])
plt.xlabel('Frequencies')
plt.ylabel('Amplitude')
plt.title('Frequency-Domain Signal')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
yf = rfft(y)
xf = rfftfreq(N, 1 / sampling_rate_fs)
filtered1 = rfft(filtered_data1)

plt.figure(figsize=(10,6))
plt.plot(xf, np.abs(yf))
plt.plot(xf, np.abs(filtered1))
plt.xlim(0, 80)
plt.xlabel('Frequencies')
plt.ylabel('Amplitude')
plt.title('Frequency-Domain Signal')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

#### High pass filter

Threshold = 30

```python
high_cutoff = 30

normal_high_cutoff = high_cutoff / nyq

b, a = butter(order, normal_high_cutoff, btype='highpass')

filtered_data2 = filtfilt(b, a, y)

plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Original Signal')
plt.plot(t, filtered_data2, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('High-Pass Filtering')
plt.legend()
plt.grid()
plt.show()
```

```python
yf = rfft(y)
xf = rfftfreq(N, 1 / sampling_rate_fs)
filtered2 = rfft(filtered_data2)

plt.figure(figsize=(10,6))
plt.plot(xf, np.abs(yf))
plt.plot(xf, np.abs(filtered2))
plt.xlim(0, 80)
plt.xlabel('Frequencies')
plt.ylabel('Amplitude')
plt.title('Frequency-Domain Signal')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

#### Bandpass Filter

Band pass range = (15, 35)

```python
low = 15
high = 35

order = 4
nyq = 0.5 * sampling_rate_fs

low_cutoff = low / nyq
high_cutoff = high / nyq

b, a = butter(4, [low_cutoff, high_cutoff], btype='band')

filtered_data3 = filtfilt(b, a, y)

plt.figure(figsize=(10, 6))
plt.plot(t, y, label='Original Signal')
plt.plot(t, filtered_data3, label='Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Band-Pass Filtering')
plt.legend()
plt.grid()
plt.show()
```

```python
yf = rfft(y)
xf = rfftfreq(N, 1 / sampling_rate_fs)
filtered3 = rfft(filtered_data3)

plt.figure(figsize=(10,6))
plt.plot(xf, np.abs(yf))
plt.plot(xf, np.abs(filtered3))
plt.xlim(0, 80)
plt.xlabel('Frequencies')
plt.ylabel('Amplitude')
plt.title('Frequency-Domain Signal')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000
duration = 1.0
t = np.linspace(0, 1, int(fs*duration), endpoint=False)
```

Synthetic signal in time-domain with frequency of 5hz, 10hz and 20hz

```python
f0 = 5
f1 = 10
f2 = 20
signal = (np.sin(2*np.pi*f0*t) + 0.2*np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t))
plt.plot(signal)
plt.show()
```

Convert signal from time-domain to frequency domain

```python
N = len(signal)
X = np.fft.fft(signal)
```

```python
freq = np.fft.fftfreq(N, 1/fs)
positive_freq = freq[:N//2]
magnitude = np.abs(X[:N//2]) * (2 / N)
```

```python
plt.figure(figsize=(10, 4))
plt.bar(positive_freq, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Frequency-Domain Representation (DFT)")
plt.show()
```

By converting from time-domain to frequency domain, all present frequencies in signal are visible in this above plot(Frequency vs Magnitude).

Reconstruct the original signal from this DFT signal

```python
X_reconstructed = np.fft.ifft(X)
X_reconstructed = np.real(X_reconstructed)
```

```python
plt.figure(figsize=(10,8))
plt.plot(t, signal, label="Original Signal")
plt.plot(t, X_reconstructed, "--", label="Reconstructed Signal")
plt.title("Original and Reconstructed signal")
plt.legend()
plt.xlabel("Time (t)")
plt.ylabel("Amplitude (A)")
plt.show()
```
